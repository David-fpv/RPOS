### **Задание 5.1**

```python
while(True):
  flag = 1
  ip = str(input())
  octets = ip.split(".")

  if len(octets) != 4:
    print('Incorrect IPv4 addres')
    flag = 0
    continue

  for octet in octets:
    if not octet.isdigit() or int(octet) > 255 or int(octet) < 0:
      print('Incorrect IPv4 addres')
      flag = 0
      break

  if flag == 1:
    if int(octets[0]) <= 223:
      print('unicast')
    elif int(octets[0]) > 223 and int(octets[0]) <= 239:
      print('multicast')
    elif int(octets[0]) == int(octets[1]) == int(octets[2]) == int(octets[3]) == 255:
      print('local broadcast')
    elif int(octets[0]) == int(octets[1]) == int(octets[2]) == int(octets[3]) == 0:
      print('unassigned')
    else:
      print('unused')
    break

```

### **Задание 5.3**

```python
access_template = ['switchport mode access',
                   'switchport access vlan',
                   'spanning-tree portfast',
                   'spanning-tree bpduguard enable']

trunk_template = ['switchport trunk encapsulation dot1q',
                  'switchport mode trunk',
                  'switchport trunk allowed vlan']

fast_int = {'access':{'0/12':'10','0/14':'11','0/16':'17','0/17':'150'},
            'trunk':{'0/1':['add','10','20'],
                     '0/2':['only','11','30'],
                     '0/4':['del','17']} }

for intf, vlan in fast_int['access'].items():
  print('interface FastEthernet' + intf)
  for command in access_template:
    if command.endswith('access vlan'):
      print(' {} {}'.format(command, vlan))
    else:
      print(' {}'.format(command))

print()

def row_with_comma (row):
  answer = ''
  for i in range(1, len(row)):
    if i != len(row) - 1:
      answer += row[i] + ','
    else:
      answer += row[i]
  return answer


for interface, config_list in fast_int['trunk'].items():
  print('interface FastEthernet' + interface)
  for command in trunk_template:
    if command.endswith('allowed vlan'):
      if config_list[0] == 'add':
        print(' {} {} {}'.format(command, 'add', row_with_comma(config_list)))
      elif config_list[0] == 'del':
        print(' {} {} {}'.format(command, 'remove', row_with_comma(config_list)))
      elif config_list[0] == 'only':
        print(' {} {}'.format(command, row_with_comma(config_list)))
    else:
      print(' {}'.format(command))

```
